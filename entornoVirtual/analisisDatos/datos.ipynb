```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Carga de datos
df = pd.read_csv("dataCGP.csv")
df
```

```python
#Total de ventas segun origen (opcion 1)

#Calcular total ventas online

total_online = df[df.Origen.isin(["Online"])]
ventas_online = sum(total_online["Total"])
print("El total de ventas online es: ", ventas_online)
```

```python
#Total de ventas segun origen (opcion 2)

#Calcular total ventas Tienda Física

total_fisica = df[df.Origen.isin(["Tienda Física"])]
ventas_fisica = sum(total_fisica["Total"])
print("El total de ventas de tienda fisica es: ", ventas_fisica)
```

```python
#Grafico de barras ventas segun origen

x = ["Online", "V_Física"]
y = [ventas_online, ventas_fisica]
fig,ax = plt.subplots()
ax.bar(x,y, linewidth=4)
fig.savefig("ventas.png")
```

```python
#Opcion 2: group by

total_ventasOrigen = df[["Origen", "Total"]]
total_ventasOrigen2 = total_ventasOrigen.groupby(by=["Origen"]).sum()
total_ventasOrigen2
```

```python
#Cantidad de ventas segun el origen

total_ventasOrigen3 = df[["Origen", "Total"]]
total_ventasOrigen4 = total_ventasOrigen3.groupby(by=["Origen"]).count()
total_ventasOrigen4
```

```python
#Grafico de barras cantidad de ventas segun origen

color = ["#95fab9", "#000000"]
x = df["Origen"].unique()
y = total_ventasOrigen4.Total
fig,ax = plt.subplots()
ax.bar(x,y, linewidth = 5, color=color)
fig.savefig("cantidadVentas.png")
```
