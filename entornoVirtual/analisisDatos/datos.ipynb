```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Para cargar los datos
df = pd.read_csv("dataCGP.csv")
df
```

```python
#Total de ventas segun origen (opcion 1)
#Calcular total ventas online
total_online = df[df.Origen.isin(["Online"])]
ventas_online = sum(total_online["Total"])
print ("ventas_online", ventas_online)
```

```python
total_online = df[df.Origen.isin(["Tienda Física"])]
ventas_fisica = sum(total_online["Total"])
print ("ventas_fisica", ventas_fisica)
```

```python
#Grafico de ventas segun origen
x = ["Online", "V_ffisica"]
y = [ventas_online, ventas_fisica]
fig,ax = plt.subplots()
ax.bar(x,y, linewidth = 4)
fig.savefig("ventas.png")
```

```python
#opcion 2: groupby
total_ventasOrigen = df [["Origen", "Total"]]
total_ventasOrigen2 = total_ventasOrigen.groupby(by=["Origen"]).sum()
total_ventasOrigen2
```
